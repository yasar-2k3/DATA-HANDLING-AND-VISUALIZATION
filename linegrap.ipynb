```python
import numpy as np
x=np.array(["a","b","c","d"])
y=np.array([20,30,40,50])
plt.scatter(x,y)
plt.show()
```

```python
#usingalpha
import numpy as np
x=np.array(["a","b","c","d"])
y=np.array([20,30,40,50])
c=["r","b","y","g"]
sizes=[100,200,300,400]
plt.scatter(x,y,color=c,s=sizes,alpha=0.5,marker="*",edgecolor="y")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```
