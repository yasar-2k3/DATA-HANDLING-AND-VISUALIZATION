```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from([1,2,3,4,5,6,])
```

```python
G.add_edges_from([(1,2),(1,3),(2,3),(3,4),(5,1),(6,1)])
```

```python
pos=nx.spring_layout(G)
```

```python
nx.draw(G,pos,with_labels=True,node_color='Green',node_size=1500,edge_color='black',linewidths=1,font_size=20)
```
